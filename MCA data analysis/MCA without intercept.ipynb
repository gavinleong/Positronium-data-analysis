```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.optimize import curve_fit # we're importing just this one function from scipy.optimize
from scipy.interpolate import UnivariateSpline
```

```python
plt.figure(num=None, figsize=(15, 8), dpi=80, facecolor='w', edgecolor='k')

ydata = np.loadtxt('N:/Lab 3/Positronium/MCA data analysis/161020_MCA_13-14psi_delay_265(5)ns.csv', comments = '#', unpack = True)
xdata = np.arange(0,2048)

index = [item for item in range(len(ydata)) if ydata[item] == 0]
ydata = [i for j, i in enumerate(ydata) if j not in index]
xdata = [i for j, i in enumerate(xdata) if j not in index]

print(np.max(ydata))

x_index = np.arange(len(xdata))


for i in x_index:
    if ydata[i] == np.max(ydata):
        start = i
        x0 = xdata[start]
print(x0)

plt.semilogy(xdata,ydata,'b.')
```

BEFORE 71 PSI MEASUREMENT (1000 nanosecond range)

```python
plt.figure(num=None, figsize=(15, 8), dpi=80, facecolor='w', edgecolor='k')

ydata1 = np.loadtxt('N:/Lab 3/Positronium/MCA data analysis/161027_MCA_14psi_delay115(2.5)ns.csv', comments = '#', unpack = True)
ydata2 = np.loadtxt('N:/Lab 3/Positronium/MCA data analysis/161027_MCA_14psi_delay162.5(2.5)ns.csv', comments = '#', unpack = True)
ydata3 = np.loadtxt('N:/Lab 3/Positronium/MCA data analysis/161027_MCA_14psi_delay265(5)ns.csv', comments = '#', unpack = True)
ydata4 = np.loadtxt('N:/Lab 3/Positronium/MCA data analysis/161027_MCA_14psi_delay540(10)ns.csv', comments = '#', unpack = True)

xdata = np.arange(1,2049) - 0.5


print(np.max(ydata1))
print(np.max(ydata2))
print(np.max(ydata3))
print(np.max(ydata4))

x_index = np.arange(len(xdata))

for i in x_index:
    if ydata1[i] == np.max(ydata1):
        start = i
        x01 = xdata[start]
        
for i in x_index:
    if ydata2[i] == np.max(ydata2):
        start = i
        x02 = xdata[start]
        
for i in x_index:
    if ydata3[i] == np.max(ydata3):
        start = i
        x03 = xdata[start]
        
for i in x_index:
    if ydata4[i] == np.max(ydata4):
        start = i
        x04 = xdata[start]

print(x01)
print(x02)
print(x03)
print(x04)

plt.figure(0)

plt.semilogy(xdata,ydata1,'b.')
plt.semilogy(xdata,ydata2,'r.')
plt.semilogy(xdata,ydata3,'g.')
plt.semilogy(xdata,ydata4,'k.')

plt.figure(1)

x0 = np.array([x01, x02, x03, x04])
y0 = np.array([115,162.5,265,540])
y0_err = np.array([2.5,2.5,5,10])
x0_err = np.array([0.5,0.5,0.5,0.5])

plt.errorbar(x0,y0,xerr = x0_err, yerr = y0_err, fmt='k.')

def line(x, m):
    '''Definition of straight line with slope m'''
    y = m*x
    return y

m = 1000 / 2048.0

print ('m', m)
guess = [m]

popt, pcov = curve_fit(line, x0, y0, p0=guess, sigma = y0_err)

slope = popt[0]
slope_err = np.sqrt(np.diag(pcov)[0])


print("error on slope", slope_err)

print("slope", slope)


x = np.linspace(0, 2048, 10000)
y = line(x, slope)

plt.plot(x,y, 'r-')

TAC_range = line(2048, slope)
TAC_range_err = 2048*slope_err
print (TAC_range)
print (TAC_range_err)
```
