```python
# used for previous own fitting code, not the script method
# defining a Positronium function
def Positronium(x,alpha,beta,sigma, h1, h2, h3):
    '''Returns a 1D array of counts for an input array of bins x.
    Functions:
    
    - gaussian 
      input parameters: 
      - standard deviation of distribution: sigma
      - peak height: h
      
    - p_decay
      input parameters:
      - free positrons decay constant: alpha
      
    - oPs_decay
      input parameters:
      - ortho-positronium decay constant: beta
      
    Other parameters:
    - background: y0
    
    Global variable:
    - peak position: x0
    '''
    
    gaussian = h1 * np.exp(-(x-x0)**2/(2*sigma**2))
    #lorentzian = (h / (1+(((x-x0)/b)**2.0)))
    p_decay = h2 * np.exp(-alpha*(x-x0))
    oPs_decay = h3 * np.exp(-beta*(x-x0))
    
    f = p_decay + oPs_decay + gaussian + y0
    return f

# h is not the true height of any of the distributions- an individual height for each dist. is 
# required
```

```python
popt, pcov = curve_fit(Positronium, xdata_cut, ydata_cut, p0 = guess)
print popt
print 'Positron lifetime', 1 / popt[0], 'ns'
print 'Positronium lifetime', 1 / popt[1], 'ns'

# ref: http://journals.aps.org/pr/pdf/10.1103/PhysRev.140.A8
# using 4.5 -+ 1.5 per microsecond per atm for positron's decay constant
# and intrinsic decay constant plus (0.27 per microsecond per atmosphere) for positronium decay constant
# Positron lifetime : 222ns
# Positronium lifetime : 137ns

y = Positronium(xdata_cut,popt[0],popt[1],popt[2],popt[3],popt[4],popt[5])

plt.figure(num=None, figsize=(15, 10), dpi=80, facecolor='w', edgecolor='k')
    
plt.plot(xdata_cut, y, 'r-', label='Least square fit')
plt.plot(xdata_cut, ydata_cut, 'b.', label='Experimental data')
plt.legend()
plt.title("Counts vs time (zoom)")
plt.xlim(120, 500)
plt.ylim(1000, 50000)

print 'Positron lifetime uncertainty', np.sqrt(np.diag(pcov)[1]), 'ns'
print 'Positronium lifetime uncertainty', np.sqrt(np.diag(pcov)[2]), 'ns'

#v.low but it doesn't take the uncertainty of the time axis

#should get:
# 0.0073 decay constant (136ns lifetime) from oPS; got 0.0107 (93ns lifetime); 0.0136 (73ns lifetime) without shifting peak
# 0.0045 decay constant (222ns lifetime) from p; got 0.9317 (1ns); 1.1957 (0.8 ns) without shifting peak
```
